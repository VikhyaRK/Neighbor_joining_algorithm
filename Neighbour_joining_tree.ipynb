```python
import numpy as np
def merge_neighbors(distance_matrix, nodes):
    current_nodes=nodes[:]
    print(current_nodes)
    sums = []
    for i in range(len(distance_matrix)):
        row_sum = 0
        for j in range(len(distance_matrix[i])):
            row_sum += distance_matrix[i][j]
        sums.append(row_sum)
        
    L=len(nodes) 
    adjusted_dm=np.zeros((L,L))
    for i in range(L):
        for j in range(L):
            adjusted_dm[i][j]=(L-2)*distance_matrix[i][j]-sums[i]-sums[j]
    for i in range(L):
        adjusted_dm[i][i]=0
        
    print(adjusted_dm)
    minimum=np.min(adjusted_dm) 
    row=np.argwhere(adjusted_dm == minimum)[0][0] 
    column=np.argwhere(adjusted_dm == minimum)[0][1] 
    

    print( nodes[row] + " and " + nodes[column]+ " are merged")
    node_a=nodes[row]  
    node_b=nodes[column] 
    merged_node=nodes[row] + nodes[column]   
    del nodes[column] 
    del nodes[row] 
    nodes=[merged_node]+nodes
    L=len(nodes)

    
    d_r=0.5*(distance_matrix[row][column]+distance_matrix[row][column+1]-distance_matrix[column][column+1])
    d_c=distance_matrix[row][column]-d_r
    print(f"Distance of {node_a} from {merged_node} = {d_r}")
    print(f"Distance of {node_b} from {merged_node} = {d_c}")


    L=len(nodes)
    to_delete=[row,column]
    temp_d_row= np.delete(distance_matrix, to_delete, axis=0)
    temp_d_column=np.delete(temp_d_row, to_delete, axis=1)
    initial_new_dm=temp_d_column
    initial_new_dm=initial_new_dm.tolist()
    
    
    

    initial_new_dm = [[0] + row for row in initial_new_dm]
    

    initial_new_dm.insert(0, [0, 0, 0, 0])
    
    
    for j in range(1,L):
        for q in range(len(current_nodes)):
            if (nodes[j]==current_nodes[q]):
                print(current_nodes[q],distance_matrix[row][q],q)
                initial_new_dm[0][j]=initial_new_dm[0][j]+distance_matrix[row][q]-d_r
                initial_new_dm[j][0]=initial_new_dm[0][j]

    print(initial_new_dm)
    final_distance_matrix=initial_new_dm
    
    return final_distance_matrix,nodes
    

    
x1,x2=merge_neighbors(dm,nodes)

y1,y2=merge_neighbors(x1,x2)
```
